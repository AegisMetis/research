Visualización de la Esperanza de vida frente a la Renta per capita para determinar si existe correlación entre estas variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Proyectos-FJCH\Udemy\data-science-con-python-numpy-pandas\data\1.3\info_pais.csv', encoding='ISO-8859-15', delimiter=';')
```

```python
df_esperanza = df.sort_values('Esperanza de vida') # Por defecto ascending=True
```

```python
df_esperanza
```

## Visualización

```python
plt.scatter(df_esperanza["Renta per capita"], df_esperanza["Esperanza de vida"]); # Valores x e y a reprentar
```

Peronalización del gráfico.

```python
plt.scatter(df_esperanza["Renta per capita"], df_esperanza["Esperanza de vida"]) # Valores x e y a reprentar
plt.title("Renta per capita vs. Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida"); # ';' Tras la última instrucción para evitar los textos descriptivos que genera matplotlib
```

```python
# Crearemos una nueva clolumna para normalizar el valor máximo de la columna población.
df_esperanza['Pobl_norm'] = df_esperanza['Poblacion'] / (max(df_esperanza['Poblacion']) / 10000)
# Dividimos el valor máximo de población por 10.000 para evitar que países como China o India 
```

```python
df_esperanza
```

```python
# Valores x e y a reprentar, s índice para la representación y c para apliicar color en función del valor de la columna indicada
plt.scatter(df_esperanza["Renta per capita"], df_esperanza["Esperanza de vida"], s=df_esperanza["Pobl_norm"], c=df_esperanza["Pobl_norm"])
plt.title("Renta per capita vs. Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(14.5, 10); # ';' Tras la última instrucción para evitar los textos descriptivos que genera matplotlib

# Para no representar los valores en las etiquetas del eje Y, le indicamos que del 1 al 120 marque una etiqueta cada 10 valores 
plt.yticks(ticks=range(1,120,10)); # ';' Tras la última instrucción para evitar los textos descriptivos que genera matplotlib

# Anotar el nombre de los 10 paíse con mayor ezperanza
for i in range(1,10):
  plt.annotate(df_esperanza["País"][i], (df_esperanza["Renta per capita"][i], df_esperanza["Esperanza de vida"][i]))
 
```

## Conclusión.
Esiste una correlación entre la Esperanza de vida y la Renta per capita ya que, como vemos en el gráfico, según aumenta esta la Renta la Esperanza de vida es mayor y no la población no es un factor determinante ya que existen países con poca o mucha población y sin embargo no se encuentran en la parte alta de la Esperanza de vida.
